# Data Visualization using Seaborn

### What is Seaborn?
- Python data visualization library based on matplotlib.
- statistical plotting library
- designed to work very well with Pandas dataframe objects.

- pip install seaborn
- conda install seaborn

- Seaborn helps resolve the two major problems faced by Matplotlib; the problems are âˆ’

>- Default Matplotlib parameters
>- Working with data frames

```python
import seaborn as sb
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = sb.load_dataset('tips')
print(df.head())
```

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5): 
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
plt.show()
```

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5):
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sb.set()
sinplot()
plt.show()
```

```python
sb.set_style("white")
sinplot()
sb.despine()
plt.show()
```

PairGrid allows us to draw a grid of subplots using the same plot type to visualize data.

```python
df = sb.load_dataset('iris')
g = sb.PairGrid(df,hue = 'species')
g.map(plt.scatter);
plt.show()
```

```python
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df,kind = 'hex')
plt.show()
```
